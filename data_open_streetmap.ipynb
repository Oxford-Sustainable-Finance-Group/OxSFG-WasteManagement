```python
import osmnx as ox
import geopandas as gpd
import pandas as pd
```

```python
place_name = "beijing, China"#"Gujarat"
# List key-value pairs for tags
tags = {'man_made': 'wastewater_plant', "name":"Waste water treatment plant"}   

buildings = ox.geometries_from_place(place_name, tags)
```

```python
place_name = "Gujarat, India"#"Gujarat"
tags = {'man_made': 'pulp and paper', "name":"pulp and paper", "industrial":"factory"}   
buildings = ox.geometries_from_place(place_name, tags)
```

**OSM Building footprints**

It is also possible to retrieve other types of OSM data features with OSMnx such as buildings or points of interest (POIs). Download the buildings with ox.geometries_from_place docs function.

When fetching specific types of geometries from OpenStreetMap using OSMnx geometries_from_place we also need to specify the correct tags. For getting all types of buildings, we can use the tag **building=yes**.
